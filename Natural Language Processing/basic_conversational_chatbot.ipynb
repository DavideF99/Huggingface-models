## Build the chatbot pipeline using ðŸ¤— Transformers Library


This code suppresses warning messages and imports the pipeline object


```python
from transformers.utils import logging
logging.set_verbosity_error()
from transformers import pipeline
```

```python
chatbot = pipeline(task='text2text-generation',
                   model="facebook/blenderbot-400M-distill")
```

```python
user_message = """
What are some fun activities I can do in the winter?
"""
```

```python
response = chatbot(user_message)
print(response)
```

- You can continue the conversation with the chatbot by creating a custom dialogue context:


```python
history = []  # this will store alternating turns


def chat(user_message):
    # Add user message to history
    history.append(f"User: {user_message}")

    # Combine all previous turns into a single prompt
    prompt = "\n".join(history)

    # Generate model response
    response = chatbot(prompt)[0]["generated_text"]

    # Add model response to history
    history.append(f"Bot: {response}")

    return response

print(chat("What are some fun activities I can do in the summer?"))
print(chat("Which ones are good for doing with friends?"))
print(chat("Which ones are cheap?"))
```
